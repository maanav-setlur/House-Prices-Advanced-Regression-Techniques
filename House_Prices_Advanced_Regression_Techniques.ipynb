# Dealing with Missing Values

```python
#Dropping columns that that have more than 75% missing values
train.dropna(axis=1, thresh=int(0.75*len(train)),inplace=True)
train.info()
```

```python
#Doing the same with missing data(2)
fillna_list=pd.DataFrame(train.isnull().any())
```

```python
fillna_list.reset_index()
```

```python
fillna_list1=list(fillna_list[fillna_list[0]==True].index)
fillna_list1
```

```python
nb_list=['LotFrontage','MasVnrArea','GarageYrBlt']
str_list1=set(fillna_list1)^set(nb_list)
str_list2=list(str_list1)
str_list2
```

```python
#numrical_fillna
for i in nb_list:
    train[i]=train[i].fillna(value=train[i].mean())
```

```python
#str_fillna
for i in str_list2:
    train[i]=train[i].fillna(value='None')
```

# feature_select

```python
train_feature=pd.concat(objs=[new_train,dummies_object],axis=1)
train_feature
```

```python
test_feature=pd.concat(objs=[test,dummies_object1],axis=1)
test_feature
```

```python
#find the feature in both train and test
both_feature=list(set(train_feature)&set(test_feature))
both_feature
```

```python
new_train_feature=train_feature[both_feature]
new_test_feature=test_feature[both_feature]
```

```python
new_train_feature
```

```python
new_test_feature
```

```python
#standard data, make predict more accuracy
from sklearn.preprocessing import StandardScaler
#write standard rule
scaler=StandardScaler().fit(new_train_feature)
#use standard rule
Std_train_feature=scaler.transform(new_train_feature)
Std_train_feature
```

```python
scaler1=StandardScaler().fit(new_test_feature)
Std_test_feature=scaler1.transform(new_test_feature)
Std_test_feature
```

```python
#split the train dataset
from sklearn.model_selection import train_test_split
x_train,y_train,x_target,y_target=train_test_split(Std_train_feature,train_target,test_size=0.3)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
tree_parameters = {"criterion":["mse", "friedman_mse",'mae'], "splitter":["best", "random"], "min_samples_split":[4,5,6], "max_depth":[2,4,6]}
tree = DecisionTreeRegressor()
clf1 = GridSearchCV(tree, tree_parameters, cv=5)
clf1.fit(x_train, x_target)
clf1.best_params_
```

```python
train_tree=DecisionTreeRegressor(criterion='mae',splitter='best',max_depth=4,min_samples_split=5)
train_tree.fit(x_train,x_target)
predict_tree=train_tree.predict(y_train)
predict_tree
```

```python
from sklearn.metrics import mean_squared_error, r2_score
# Explained variance score: 1 is perfect prediction 
#  R2 决定系数（拟合优度）
# 模型越好：r2→1
# 模型越差：r2→0
print('Variance score: %.2f' % r2_score(y_target, predict_tree))
```
